```python
import ast

def convert(string_list):
    l = []
    for i in ast.literal_eval(string_list): 
                                                    #The coming paramter was string list and we have to iterate over list so first 
                                                    #we convert the string list -> List
        l.append(i['name'])                         
        
    return l
```

# stemming

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)

# stemming only one row
# string-> converted to list-> applied loop-> stemmed each word and appended into new list-> again converted list back to string
```
